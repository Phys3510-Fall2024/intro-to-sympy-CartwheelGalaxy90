```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
F_0 = 1.0 # Amplitude of the driving force
m = 1.0 # Mass of the system
omega_0 = 1.0 # Natural frequency
gamma = 0.1 # Damping coefficient

# Frequency range
omega = np.linspace(0, 2 * omega_0, 500)

# Calculate C_bar and phi
C_bar = (F_0 / m) / np.sqrt((omega_0**2 - omega**2)**2 + 4 * gamma**2 * omega**2)
phi = np.arctan2(2 * gamma * omega, omega_0**2 - omega**2)

# Plot C_bar
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(omega, C_bar, label=r'')
plt.xlabel(r'')
plt.ylabel(r'')
plt.title('Frequency vs Amplitude')
plt.grid(True)

# Plot phi
plt.subplot(1, 2, 2)
plt.plot(omega, phi, label=r'', color='orange')
plt.xlabel(r'')
plt.ylabel(r' (radians)')
plt.title('Frequency vs Phase')
plt.grid(True)

plt.tight_layout()
plt.show()
```
